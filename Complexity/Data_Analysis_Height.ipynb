We have therefore shown that the standard deviation of the slope will tend to zero as $L \rightarrow \infty$

```python
def probability(list_of_heights):
    """
    returns normalised probablity for each unique value of the input array
    """
    a = Counter(list_of_heights)
    height =[]
    f = []
    for i in a:
        height.append(i)
        f.append(a[i])
        N = np.sum(f)
    return height, f/N
```
